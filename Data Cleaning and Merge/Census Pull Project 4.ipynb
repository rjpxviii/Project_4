```python
pov = pd.read_csv("Resources/ACSST5Y2022.S1701-Data.csv")
pov
```

```python
filter = pov["NAME"].str.contains("CDP")
filtered_pov = pov[~filter]
filtered_pov["NAME"] = filtered_pov["NAME"].str.replace('city, Texas', '')
filtered_pov["NAME"] = filtered_pov["NAME"].str.replace('town, Texas', '')
filtered_pov
```

```python
poverty = pd.DataFrame({
    "City" : filtered_pov["NAME"],
    "Percent_Male" : filtered_pov["S1701_C03_011E"],
    "Percent_Female": filtered_pov["S1701_C03_012E"],
    "Percent_over_25y" : filtered_pov["S1701_C03_022E"],
    "Percent_HS_less" : filtered_pov["S1701_C03_023E"],
    "Percent_HS": filtered_pov["S1701_C03_024E"],
    "Percent_Some_Col_Assoc" : filtered_pov["S1701_C03_025E"],
    "Percent_BS_over" : filtered_pov["S1701_C03_025E"],
    "Percent_unemployed": filtered_pov["S1701_C03_031E"],
    "Population" : filtered_pov["S1701_C01_001E"],
    "Population_Below_Poverty": filtered_pov["S1701_C01_038E"],                               
})
poverty
```

```python
poverty.info()
```

```python
poverty = poverty.drop(0)
poverty
```

```python
poverty["Population"] = poverty["Population"].astype(int)
poverty["Population_Below_Poverty"]=poverty["Population_Below_Poverty"].astype(int)
poverty["Poverty_Rate"] = (poverty["Population_Below_Poverty"] / poverty["Population"]) * 100
poverty
```
